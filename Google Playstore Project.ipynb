# 8. Convert the column "Reviews'' to numeric data type and check the presence of outliers in the column and handle the outliers using a transformation approach

```python
df['Reviews']=pd.to_numeric(df['Reviews'], errors='coerce')
df['Reviews']
```

```python
Q1=df['Reviews'].quantile(0.25)
Q3=df["Reviews"].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
outliers= df[(df['Reviews'] < lower_bound) | (df['Reviews'] > upper_bound)]
outliers
```

# 9. The column 'Size' contains alphanumeric values, treat the non numeric data and convert the column into suitable data type. (hint: Replace M with 1 million and K with 1 thousand, and drop the entries where size='Varies with device') 

```python
df['Size']
```

```python
df['Size']= df['Size'].replace('Varies with decvice', np.nan)
df['Size']=df['Size'].str.replace('M','e6').str.replace('K','e3')
df["Size"]=pd.to_numeric(df["Size"],errors="coerce")
df=df.dropna(subset=['Size'])
```

```python
df
```

# 10. Check the column 'Installs', treat the unwanted characters and convert the column into a suitable data type

```python
df['Installs'].info()
```

```python
df["Installs"]=df['Installs'].str.replace(',','').str.replace('+','')
df['Installs']=pd.to_numeric(df['Installs'],errors="coerce")
df
```
